# Statistics Advance 2 Theory

1. What is hypothesis testing in statistics?
  - Hypothesis testing in statistics is a method used to make decisions or inferences about a population based on a sample of data.

2. What is the null hypothesis, and how does it differ from the alternative hypothesis?
  - Null hypothesis :- The null hypothesis is a statement that there is no effect, no difference, or nothing unusual happening.
It assumes that things are just normal or as expected.
  - Example:
If a company claims its battery lasts 10 hours, the null hypothesis would be:
H‚ÇÄ: Battery life = 10 hours
  - Alternative Hypothesis :- The alternative hypothesis is what you want to prove.
It states that there is an effect, a difference, or a change.
  - Example:
If you believe the battery lasts less than 10 hours, the alternative hypothesis would be:H‚ÇÅ: Battery life < 10 hours

3. What is the significance level in hypothesis testing, and why is it important.
  - The significance level, usually written as Œ± (alpha), is a threshold we set in hypothesis testing to decide how much risk we are willing to take of making a wrong decision ‚Äî specifically, rejecting the null hypothesis when it is actually true.
  - Why is it Important?
      * Because it helps you decide when to reject the null hypothesis (H‚ÇÄ).
      * If the p-value ‚â§ Œ±, you reject H‚ÇÄ ‚Üí your result is statistically significant.
      * If the p-value > Œ±, you do not reject H‚ÇÄ ‚Üí not enough evidence to support the alternative.

4. What does a P-value represent in hypothesis testing?
  - The p-value is a number that helps you decide whether to reject the null hypothesis (H‚ÇÄ).

5.  How do you interpret the P-value in hypothesis testing?
  - The p-value helps you decide whether your sample results are strong enough to reject the null hypothesis (H‚ÇÄ).

6. What are Type 1 and Type 2 errors in hypothesis testing?
  - Type 1 Error (False Positive)
      * This happens when you reject the null hypothesis (H‚ÇÄ) even though it is true.
      * You're saying there is an effect or difference, but actually there isn't.
  - Type 2 Error (False Negative)
      * This happens when you do not reject the null hypothesis even though it is false.
      * You're saying there is no effect, but actually there is.

7. What is the difference between a one-tailed and a two-tailed test in hypothesis testing?
  - One-Tailed Test :- A one-tailed test checks for an effect in only one direction ‚Äî either greater than or less than a certain value.
  - wo-Tailed Test:- A two-tailed test checks for an effect in both directions either greater than or less than a certain value.

8. What is the Z-test, and when is it used in hypothesis testing?
  - The Z-test is a type of statistical test used in hypothesis testing when you want to compare a sample mean to a known population mean and the population standard deviation is known.
  - When to Use a Z-test:
      * The sample size is large (usually n ‚â• 30)
      * The population standard deviation (œÉ) is known
      * The data is normally distributed, or the sample size is large enough for the Central Limit Theorem to apply.

9. How do you calculate the Z-score, and what does it represent in hypothesis testing?
  - The Z-score tells you how many standard deviations your sample mean (or data point) is away from the population mean.
  - What the Z-score Represents:
      * The Z-score measures how far (in terms of standard deviations) your sample result is from the null hypothesis mean.
      * A high absolute value of Z means your sample mean is far from the population mean, indicating the data may not support the null hypothesis.
      * A low absolute value means the sample mean is close to the population mean, so data supports the null hypothesis.

10. What is the T-distribution, and when should it be used instead of the normal distribution?
  - The T-distribution is a probability distribution similar to the normal distribution but with heavier tails. This means it gives more probability to values far from the mean. It looks like a bell curve but is wider and shorter, especially with smaller sample sizes.
  - Use the T-distribution instead of the normal distribution when:
      * The population standard deviation (œÉ) is unknown.
      * The sample size is small (usually n < 30).
      * The data is approximately normally distributed (or close enough).

11.  What is the difference between a Z-test and a T-test?
  - Z-test:
      * Used when the population standard deviation (œÉ) is known.
      * Usually applied with large sample sizes (n ‚â• 30).
      * Based on the normal distribution (Z-distribution).
      * Tests whether a sample mean differs from the population mean.
  - T-test:
      * Used when the population standard deviation (œÉ) is unknown and estimated from the sample.
      * Typically used with small sample sizes (n < 30).
      * Based on the t-distribution, which accounts for extra uncertainty.
      * Tests whether a sample mean differs from the population mean (one-sample), or compares means of two groups (two-sample).

12. What is the T-test, and how is it used in hypothesis testing?
  - The T-test is a statistical test used to determine whether there is a significant difference between the means of two groups or between a sample mean and a known value when the population standard deviation (œÉ) is unknown.
  - It uses the t-distribution, which adjusts for the extra uncertainty caused by estimating the population standard deviation from the sample.

13. What is the relationship between Z-test and T-test in hypothesis testing?
  - Both the Z-test and T-test are used to compare sample means to a population mean (or compare two means) to decide if there‚Äôs a significant difference.
  - Both tests check if sample data supports or rejects the null hypothesis about population means.
  - Z-test Used when the population standard deviation (œÉ) is known.
  - T-test Used when the population standard deviation is unknown and estimated from the sample.
  - Z-test Best for large samples (usually n ‚â• 30), where sample estimates approximate population parameters well.
  - T-test Designed for small samples (usually n < 30), where estimating œÉ adds uncertainty.

14. What is a confidence interval, and how is it used to interpret statistical results?
  - A confidence interval is a range of values, calculated from sample data, that is likely to contain the true population parameter (like the mean) with a certain level of confidence.
  - Example : a 95% confidence interval means you are 95% confident that the true population parameter lies within this range.
  
  - It Used to Interpret Statistical Results:-
      * It gives a range of plausible values for the population parameter, not just a single estimate.
      * If the confidence interval does not include the value stated in the null hypothesis (like zero difference), it suggests the result is statistically significant.
      * It provides more information than a simple yes/no decision, showing how precise or uncertain the estimate is.

15.  What is the margin of error, and how does it affect the confidence interval?
  - The margin of error (MoE) is the amount you're allowing for possible error when estimating a population value based on a sample.
  - It tells you how much the sample result might differ from the true population value due to natural variability or randomness in sampling.
  - A confidence interval is built around the sample mean like this:
      * Confidence¬†Interval=ùë•Àâ¬±Margin¬†of¬†Error

16. How is Bayes' Theorem used in statistics, and what is its significance?
  - Bayes‚Äô Theorem is a formula that helps you update your beliefs or probabilities after seeing new evidence.
It links prior knowledge with new data to find the updated probability of an event.
  - Bayes‚Äô Theorem is central to Bayesian statistics, where we:
      * Start with a prior belief (called a prior probability)
      * Collect data (new evidence)
      * Use Bayes‚Äô Theorem to update the belief ‚Üí get the posterior probability.

17. What is the Chi-square distribution, and when is it used?
  - The Chi-square (œá¬≤) distribution is a type of probability distribution that is used mainly for analyzing categorical data. It is not symmetric instead, it is skewed to the right, especially for smaller sample sizes.
  - It is used to test how well observed data fit with expected data ‚Äî basically, it tells you if there‚Äôs a significant difference between what you expect and what you actually get.

18. What is the Chi-square goodness of fit test, and how is it applied?
  - The Chi-square goodness-of-fit test is a statistical test used to determine whether a set of observed frequencies matches the expected frequencies for a single categorical variable.
  -  Steps to Apply the Test:
      * Null hypothesis (H‚ÇÄ): The observed data follows the expected distribution.
      * Alternative hypothesis (H‚ÇÅ): The observed data does not follow the expected distribution.
      * Calculate the expected frequencies based on the theoretical distribution.
      * Use the formula to calculate the Chi-square statistic.
      * Find the critical value or P-value from the Chi-square distribution table using the degrees of freedom (df = number of categories - 1).
      * Compare and conclude:If œá¬≤ is greater than the critical value (or P-value < significance level), reject the null hypothesis.

19. What is the F-distribution, and when is it used in hypothesis testing?
  - The F-distribution is a probability distribution used mainly to compare variances between two groups or models. It is positively skewed and only takes positive values.
  - It is used when the test statistic is a ratio of two variances, which is why it appears often in comparing statistical models or in analysis of variance (ANOVA).

20. What is an ANOVA test, and what are its assumptions?
  - ANOVA stands for Analysis of Variance. It is a statistical test used to determine whether there are any significant differences between the means of three or more independent groups.
  - Instead of comparing each group one by one (which increases the risk of errors), ANOVA tests them all at once.
  - Assumptions of ANOVA:
      * Independence :- Observations should be independent from each other (no overlapping samples).
      * Normality :- The data in each group should be approximately normally distributed.
      * Equal Variances:- The variances of the groups should be roughly equal.

21. What are the different types of ANOVA tests?
  - One Way Anova : Compares the means of three or more independent groups based on one factor.
  - Repeated Measures ANOVA : Used when the same subjects are measured multiple times.
  - Factorial Anova : Two or more factor each of which with atleat two levels. Levels can be either dependent, independent or both.
  
22. What is the F-test, and how does it relate to hypothesis testing?
  - The F-test is a statistical test used to compare the variances of two or more groups, or to check if a model is significantly better than another.
  - The F-test is used in hypothesis testing to decide whether to reject the null hypothesis. Here's how:
      * Null Hypothesis (H‚ÇÄ): All group variances (or means, in ANOVA) are equal ‚Äî no effect or difference.
      * Alternative Hypothesis (H‚ÇÅ): At least one group is different ‚Äî there is a significant effect.
      * Decision Rule: If the F-statistic is large and the P-value is small (less than significance level like 0.05), reject H‚ÇÄ. therwise, fail to reject H‚ÇÄ.

```python
# 3. Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution?
def bernoulli_pdf(x, p):

    if x not in [0, 1]:
        raise ValueError("x must be 0 or 1")
    if not (0 <= p <= 1):
        raise ValueError("p must be between 0 and 1")

    return (p ** x) * ((1 - p) ** (1 - x))

# Example usage
print("P(X=0):", bernoulli_pdf(0, 0.3))
print("P(X=1):", bernoulli_pdf(1, 0.3))
```

```python
# 9. Write a Python function to calculate Z-scores from a dataset and plot them?
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

def calculate_and_plot_z_scores(data):
    # Calculate mean and standard deviation
    mean = np.mean(data)
    std = np.std(data)

    # Calculate Z-scores
    z_scores = [(x - mean) / std for x in data]

    plt.figure(figsize=(10, 5))
    sns.barplot(x=list(range(len(data))), y=z_scores, palette="coolwarm")
    plt.axhline(0, color='black', linestyle='--')
    plt.title("Z-scores of the Dataset")
    plt.xlabel("Index")
    plt.ylabel("Z-score")
    plt.tight_layout()
    plt.show()

    return z_scores

data = [10, 12, 23, 23, 16, 23, 21, 16]
z_scores = calculate_and_plot_z_scores(data)
print("Z-scores:", z_scores)
```

```python
# 10. Implement the Central Limit Theorem (CLT) using Python for a non-normal distribution?
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

def demonstrate_clt(population_size=100000, sample_size=30, num_samples=1000):
    # Create a non-normal population
    population = np.random.exponential(scale=2.0, size=population_size)

    # Draw repeated samples and compute their means
    sample_means = []
    for _ in range(num_samples):
        sample = np.random.choice(population, size=sample_size, replace=False)
        sample_means.append(np.mean(sample))

    # Plotting
    plt.figure(figsize=(12, 5))

    # Plot original population
    plt.subplot(1, 2, 1)
    sns.histplot(population, bins=50, kde=True, color='skyblue')
    plt.title("Original Population (Exponential)")
    plt.xlabel("Value")
    plt.ylabel("Frequency")

    # Plot distribution of sample means
    plt.subplot(1, 2, 2)
    sns.histplot(sample_means, bins=50, kde=True, color='salmon')
    plt.title(f"Sampling Distribution of the Mean\n(n={sample_size}, samples={num_samples})")
    plt.xlabel("Sample Mean")
    plt.ylabel("Frequency")

    plt.tight_layout()
    plt.show()

demonstrate_clt(sample_size=50, num_samples=1000)
```

```python
# 14. Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal distribution.
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def calculate_z_score(x, mean, std):
    z = (x - mean) / std
    return z

def plot_z_score(z):
    # Generate standard normal curve
    x_vals = np.linspace(-4, 4, 1000)
    y_vals = norm.pdf(x_vals, 0, 1)

    plt.figure(figsize=(10, 5))
    plt.plot(x_vals, y_vals, label='Standard Normal Curve', color='blue')
    plt.fill_between(x_vals, y_vals, alpha=0.1, color='blue')
    plt.axvline(z, color='red', linestyle='--', label=f'Z-score = {z:.2f}')
    plt.title("Z-score on Standard Normal Distribution")
    plt.xlabel("Z-score")
    plt.ylabel("Probability Density")
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

def main():
    x = 75
    mean = 70
    std = 5

    z = calculate_z_score(x, mean, std)
    print(f"Z-score for x = {x} is: {z:.2f}")
    plot_z_score(z)

if __name__ == "__main__":
    main()
```

```python
# 15. Implement hypothesis testing using Z-statistics for a sample dataset?
import numpy as np
from scipy.stats import norm

def z_test(sample_data, population_mean, population_std, alpha=0.05):
   # Compute sample statistics
    sample_mean = np.mean(sample_data)
    n = len(sample_data)

    #  Calculate Z-statistic
    z_score = (sample_mean - population_mean) / (population_std / np.sqrt(n))

    #  Determine critical value for two-tailed test
    z_critical = norm.ppf(1 - alpha / 2)

    # Decision
    reject_null = abs(z_score) > z_critical

    print(f"Sample Mean: {sample_mean:.2f}")
    print(f"Z-score: {z_score:.2f}")
    print(f"Z-critical (two-tailed, Œ±={alpha}): ¬±{z_critical:.2f}")
    print("Result:", "Reject H‚ÇÄ " if reject_null else "Fail to reject H‚ÇÄ ")

    return z_score, reject_null

# Sample data (e.g., test scores)
sample = [68, 70, 72, 65, 74, 69, 71, 73, 70, 68]

# Population parameters
population_mean = 70
population_std = 4  # Known population standard deviation

# Run Z-test
z_test(sample, population_mean, population_std, alpha=0.05)
```

```python
# 16. Create a confidence interval for a dataset using Python and interpret the result.
import numpy as np
from scipy.stats import norm

def confidence_interval(data, confidence=0.95):
    n = len(data)
    mean = np.mean(data)
    std = np.std(data, ddof=1)

    # Standard error
    se = std / np.sqrt(n)

    # Z-critical value
    z_critical = norm.ppf(1 - (1 - confidence) / 2)

    # Margin of error
    margin = z_critical * se

    # Confidence interval
    lower = mean - margin
    upper = mean + margin

    print(f"Sample Mean: {mean:.2f}")
    print(f"{int(confidence*100)}% Confidence Interval: ({lower:.2f}, {upper:.2f})")
    print(f"Interpretation: We are {int(confidence*100)}% confident that the true population mean lies between {lower:.2f} and {upper:.2f}")

    return (lower, upper)

data = [68, 70, 72, 65, 74, 69, 71, 73, 70, 68]
confidence_interval(data, confidence=0.95)
```

```python
# 17. Generate data from a normal distribution, then calculate and interpret the confidence interval for its mean?
import numpy as np
from scipy.stats import norm

def generate_normal_data(mean=100, std=15, size=50):
    """Generate normally distributed data"""
    data = np.random.normal(loc=mean, scale=std, size=size)
    return data

def confidence_interval(data, confidence=0.95):
    n = len(data)
    sample_mean = np.mean(data)
    sample_std = np.std(data, ddof=1)
    se = sample_std / np.sqrt(n)

    # Get the Z critical value
    z_critical = norm.ppf(1 - (1 - confidence) / 2)

    # Margin of error and CI
    margin = z_critical * se
    lower = sample_mean - margin
    upper = sample_mean + margin

    print(f"Sample Mean: {sample_mean:.2f}")
    print(f"{int(confidence * 100)}% Confidence Interval: ({lower:.2f}, {upper:.2f})")
    print(f"Interpretation: We are {int(confidence * 100)}% confident that the true mean lies between {lower:.2f} and {upper:.2f}")

    return lower, upper

# Main
if __name__ == "__main__":
    np.random.seed(42)  # for reproducibility
    sample_data = generate_normal_data(mean=100, std=15, size=50)
    confidence_interval(sample_data, confidence=0.95)
```

```python
# 19. Use Python to calculate and interpret the cumulative distribution function (CDF) of a Poisson distribution.
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

def poisson_cdf(lmbda=3, max_k=15):

    # Discrete x values (k)
    x = np.arange(0, max_k + 1)

    # Compute the CDF values
    cdf_values = poisson.cdf(x, mu=lmbda)

    # Display values
    for k, prob in zip(x, cdf_values):
        print(f"P(X ‚â§ {k}) = {prob:.4f}")

    # Plot the CDF
    plt.figure(figsize=(10, 5))
    plt.step(x, cdf_values, where='post', color='green', label=f'Poisson CDF (Œª={lmbda})')
    plt.scatter(x, cdf_values, color='green')
    plt.title(f'Poisson Distribution CDF (Œª = {lmbda})')
    plt.xlabel('k (Number of Events)')
    plt.ylabel('Cumulative Probability')
    plt.grid(True)
    plt.legend()
    plt.tight_layout()
    plt.show()

# Example usage
if __name__ == "__main__":
    poisson_cdf(lmbda=3, max_k=15)
```

```python
# 23. Implement Z-score normalization on a dataset and visualize its transformation.

import numpy as np
import matplotlib.pyplot as plt

def z_score_normalization(data):
    mean = np.mean(data)
    std = np.std(data)
    z_scores = (data - mean) / std
    return z_scores, mean, std

def visualize_transformation(original, normalized):
    plt.figure(figsize=(12, 5))

    # Original Data
    plt.subplot(1, 2, 1)
    plt.hist(original, bins=30, color='skyblue', edgecolor='black')
    plt.title("Original Data")
    plt.xlabel("Value")
    plt.ylabel("Frequency")
    plt.axvline(np.mean(original), color='red', linestyle='--', label='Mean')
    plt.legend()

    # Normalized Data
    plt.subplot(1, 2, 2)
    plt.hist(normalized, bins=30, color='lightgreen', edgecolor='black')
    plt.title("Z-score Normalized Data")
    plt.xlabel("Z-score")
    plt.axvline(0, color='red', linestyle='--', label='Mean = 0')
    plt.legend()

    plt.tight_layout()
    plt.show()

# Example usage
if __name__ == "__main__":
    np.random.seed(0)
    data = np.random.normal(loc=50, scale=10, size=1000)

    normalized_data, mean, std = z_score_normalization(data)

    print(f"Original Mean: {mean:.2f}, Std Dev: {std:.2f}")
    print(f"Normalized Mean: {np.mean(normalized_data):.2f}, Std Dev: {np.std(normalized_data):.2f}")

    visualize_transformation(data, normalized_data)
```

```python
# 4. Perform a two-tailed Z-test using Python and visualize the decision region on a plot?
import numpy as np
import math
import matplotlib.pyplot as plt
from scipy.stats import norm

#  Define population and sample parameters
population_mean = 100
population_std_dev = 15
alpha = 0.05

np.random.seed(42)
sample_size = 50
sample_data = np.random.normal(loc=105, scale=15, size=sample_size)
sample_mean = np.mean(sample_data)

# Calculate Z-score
z_score = (sample_mean - population_mean) / (population_std_dev / math.sqrt(sample_size))

# Calculate critical Z-values for two-tailed test
z_critical = norm.ppf(1 - alpha/2)
neg_z_critical = -z_critical
# Calculate p-value ---
p_value = 2 * (1 - norm.cdf(abs(z_score)))

print(f"Sample Mean: {sample_mean:.2f}")
print(f"Z-score: {z_score:.4f}")
print(f"P-value: {p_value:.4f}")

if abs(z_score) > z_critical:
    decision = "Reject H‚ÇÄ"
    interpretation = "Sample mean is significantly different from population mean."
else:
    decision = "Fail to reject H‚ÇÄ"
    interpretation = "No significant difference."

print(f" Decision: {decision}")
print(f" Interpretation: {interpretation}")

# Visualization
x = np.linspace(-4, 4, 1000)
y = norm.pdf(x, 0, 1)

plt.figure(figsize=(10, 6))
plt.plot(x, y, label='Z-distribution')

# Shade rejection regions
plt.fill_between(x, y, where=(x < neg_z_critical), color='red', alpha=0.3, label='Rejection Region')
plt.fill_between(x, y, where=(x > z_critical), color='red', alpha=0.3)

plt.axvline(z_score, color='blue', linestyle='--', linewidth=2, label=f'Z-score = {z_score:.2f}')
plt.axvline(z_critical, color='black', linestyle=':', label=f'Z critical = ¬±{z_critical:.2f}')
plt.axvline(neg_z_critical, color='black', linestyle=':')
plt.title('Two-tailed Z-Test Visualization')
plt.xlabel('Z-value')
plt.ylabel('Probability Density')
plt.legend()
plt.show()
```

```python
# 5. Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing.
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
import math

def visualize_type1_type2_errors(mu0, mu1, sigma, n, alpha=0.05, test_type='right'):
    # Calculate standard error
    se = sigma / math.sqrt(n)

    # Calculate critical value for H0
    if test_type == 'right':
        z_critical = norm.ppf(1 - alpha)
        critical_value = mu0 + z_critical * se
        beta = norm.cdf((critical_value - mu1) / se)
    elif test_type == 'left':
        z_critical = norm.ppf(alpha)
        critical_value = mu0 + z_critical * se
        beta = 1 - norm.cdf((critical_value - mu1) / se)
    else:
        raise ValueError("test_type must be either 'right' or 'left'")

    # x-axis range
    x = np.linspace(mu0 - 4*se, mu1 + 4*se, 1000)

    # PDFs
    y_H0 = norm.pdf(x, mu0, se)
    y_H1 = norm.pdf(x, mu1, se)

    # Plot the distributions
    plt.figure(figsize=(10, 6))
    plt.plot(x, y_H0, label='H‚ÇÄ Distribution', color='blue')
    plt.plot(x, y_H1, label='H‚ÇÅ Distribution', color='green')

    # Shade Type I Error region (Œ±)
    if test_type == 'right':
        plt.fill_between(x, y_H0, where=(x > critical_value), color='red', alpha=0.3, label='Type I Error (Œ±)')
    else:
        plt.fill_between(x, y_H0, where=(x < critical_value), color='red', alpha=0.3, label='Type I Error (Œ±)')

    # Shade Type II Error region (Œ≤)
    if test_type == 'right':
        plt.fill_between(x, y_H1, where=(x < critical_value), color='orange', alpha=0.3, label='Type II Error (Œ≤)')
    else:
        plt.fill_between(x, y_H1, where=(x > critical_value), color='orange', alpha=0.3, label='Type II Error (Œ≤)')

    # Vertical line at critical value
    plt.axvline(critical_value, color='black', linestyle='--', label=f'Critical Value = {critical_value:.2f}')

    # Plot settings
    plt.title('Type I and Type II Errors in Hypothesis Testing')
    plt.xlabel('Sample Mean')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.grid(True)
    plt.show()

    # Print results
    print(f"Critical Value: {critical_value:.2f}")
    print(f"Type I Error (Œ±): {alpha}")
    print(f"Type II Error (Œ≤): {beta:.4f}")
    print(f"Power of the test (1 - Œ≤): {1 - beta:.4f}")
visualize_type1_type2_errors(mu0=100, mu1=105, sigma=15, n=50, alpha=0.05, test_type='right')
```

```python
# 9. Write a Python function to calculate the confidence interval for a sample mean and explain its significance?
import numpy as np
from scipy.stats import t

def confidence_interval(data, confidence=0.95):
    n = len(data)
    mean = np.mean(data)
    std_err = np.std(data, ddof=1) / np.sqrt(n)

    t_critical = t.ppf((1 + confidence) / 2, df=n-1)
    margin_of_error = t_critical * std_err

    lower = mean - margin_of_error
    upper = mean + margin_of_error

    return (lower, upper, mean, margin_of_error)

# Simulate sample data
np.random.seed(0)
sample_data = np.random.normal(loc=100, scale=15, size=30)

ci_lower, ci_upper, sample_mean, moe = confidence_interval(sample_data, confidence=0.95)

print(f"Sample Mean: {sample_mean:.2f}")
print(f"95% Confidence Interval: ({ci_lower:.2f}, {ci_upper:.2f})")
print(f"Margin of Error: ¬±{moe:.2f}")
```

```python
# 11. Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process.

def bayesian_update(prior, likelihood, false_positive_rate):
    # Compute the total probability of a positive test
    evidence = likelihood * prior + false_positive_rate * (1 - prior)

    # Apply Bayes' Theorem
    posterior = (likelihood * prior) / evidence
    return posterior

# Parameters
prior = 0.01               # P(Disease)
likelihood = 0.99          # P(Positive | Disease)
false_positive_rate = 0.05 # P(Positive | No Disease)

posterior = bayesian_update(prior, likelihood, false_positive_rate)

print(f"Probability of having the disease given a positive test: {posterior:.4f}")
```

```python
# 12. Perform a Chi-square test for independence between two categorical variables in Python.
import pandas as pd
import scipy.stats as stats

# Create a contingency table
# Example: Gender vs Preference
data = {
    'Gender': ['Male', 'Male', 'Female', 'Female', 'Male', 'Female', 'Male', 'Female'],
    'Preference': ['A', 'B', 'A', 'A', 'B', 'B', 'A', 'B']
}

df = pd.DataFrame(data)

# Create a contingency table
contingency_table = pd.crosstab(df['Gender'], df['Preference'])

print("Contingency Table:")
print(contingency_table)

# Perform Chi-square test
chi2_stat, p_value, dof, expected = stats.chi2_contingency(contingency_table)

print("\nChi-square Statistic:", chi2_stat)
print("Degrees of Freedom:", dof)
print("P-value:", p_value)
print("\nExpected Frequencies Table:")
print(expected)
```

```python
# 13. Write a Python program to calculate the expected frequencies for a Chi-square test based on observed data.
import numpy as np
import pandas as pd

# Create observed frequency table
# Example: Gender vs Product Preference
observed = pd.DataFrame([
    [10, 20],
    [20, 30]
], index=['Male', 'Female'], columns=['Product A', 'Product B'])

print("Observed Frequencies:")
print(observed)

# Calculate expected frequencies manually
row_totals = observed.sum(axis=1).values.reshape(-1, 1)
col_totals = observed.sum(axis=0).values.reshape(1, -1)
total = observed.values.sum()

expected = (row_totals @ col_totals) / total
expected_df = pd.DataFrame(expected, index=observed.index, columns=observed.columns)

print("\nExpected Frequencies:")
print(expected_df.round(2))
```
